### Análisis Comparativo de los Principales Índices Bursátiles Globales (2018-2024)

#### TEMA: Finanzas

### Describe brevemente el tema del proyecto en Markdown:

En este análisis, se comparan los comportamientos históricos de diversos índices bursátiles globales durante el período de 2018 a 2024. Los índices seleccionados incluyen los más representativos a nivel mundial, como el S&P 500, Nasdaq 100, Dow Jones, Russell 2000, así como índices europeos y asiáticos, como el DAX 40, CAC 40, FTSE 100, Nikkei 225, Hang Seng y Euro Stoxx 50.

El objetivo es evaluar el rendimiento, la volatilidad y las tendencias generales de cada uno, con el fin de entender cómo se han comportado los mercados en diferentes regiones del mundo durante este período de tiempo. A través de gráficos y estadísticas descriptivas, se analizarán los precios de apertura, cierre, máximo, mínimo, y el volumen de negociación, proporcionando una visión integral del desempeño de estos índices.


### Escribe tu hipótesis principal en Markdown:

1. Correlación entre índices globales: Los índices bursátiles globales están altamente correlacionados debido a factores económicos comunes.

2. Volatilidad de índices pequeños vs grandes: Los índices más pequeños muestran mayor volatilidad que los índices grandes y establecidos.

3. Resiliencia a crisis globales: Los índices de economías desarrolladas se recuperan más rápido de crisis globales que los de economías emergentes.

4. Rendimiento relativo de índices: El Nasdaq 100 ha superado al S&P 500 debido a su enfoque en empresas tecnológicas.

5. Efectos regionales en el rendimiento: Los índices europeos muestran menor rentabilidad en comparación con los índices estadounidenses.

6. Impacto de factores externos en mercados emergentes: Los índices de mercados emergentes son más sensibles a eventos económicos globales que los índices de mercados desarrollados.

7. Diferencias de rendimiento por sector: Los índices con mayor peso en el sector tecnológico han tenido un rendimiento superior en los últimos años.
